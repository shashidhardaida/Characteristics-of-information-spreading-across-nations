```
tweet_texts_to_compare = []
tweet_texts_to_compare = df['tweet_text'].tolist()
tweet_texts_to_compare
```

```
for i in range(len(tweet_texts_to_compare)):
  if type(tweet_texts_to_compare[i])==str:
    tweet_texts_to_compare[i] = tweet_texts_to_compare[i].lower()
tweet_texts_to_compare
```

```
Id,time_stamp,url,tweet_text,username,favorite_count,followers,all_hastages,geolocation =[],[],[],[],[],[],[],[],[]

for h,i in enumerate(tweet_texts_to_compare):
  for j,k in enumerate(keywords_political):
      if k in i:
        Id.append(df['ID'][h])
        time_stamp.append(df['timestamp'][h])
        url.append(df['url'][h])
        tweet_text.append(df['tweet_text'][h])
        username.append(df['username'][h])
        favorite_count.append(df['favotite_count'][h])
        followers.append(df['followers_count'][h])
        all_hastages.append(df['all_hashtags'][h])
        geolocation.append(df['geolocation'][h])
        print('Processing....')
```

```
political_data = pd.DataFrame(columns=column_names)
```

```
political_data['id'] = Id
political_data['time_stamp'] = time_stamp
political_data['tweet_text'] = tweet_text
political_data['username'] = username
political_data['url'] = url
political_data['all_hashtags'] = all_hastages
political_data['followers'] = followers
political_data['favorite_count'] = favorite_count
political_data['geolocation'] = geolocation
```

```
political_data = political_data.drop_duplicates(subset='id', keep='first')
political_data
```
